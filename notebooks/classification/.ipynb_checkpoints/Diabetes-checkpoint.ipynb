## Preprocessing data

### seprate x and y
We use `drop()` function for remove **Outcome** from **x** data and get values for both of theme.

```python
x = df.drop(['Outcome'], axis=1).values
y = df['Outcome'].values
```

### Normalize Data
We use `preprocessing` and call the `StandardScaler` function from it for `fit()` **x** and then transform it as type float.

```python
from sklearn import preprocessing

scaler = preprocessing.StandardScaler().fit(x)
x = scaler.transform(x.astype(float))
```

### Split train and test data using sklearn
We use `train_test_split` form spliting train and test 

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

## Predict and test
We use a loop for test `k` for 0-10 numbers. and check accuracy score for choice best number for `k`.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

k = 10
main_accuracy= np.zeros((k-1))

for i in range(1, k):
    neigh = KNeighborsClassifier(n_neighbors = i).fit(x_train, y_train)
    y_hat = neigh.predict(x_test)
    main_accuracy[i-1] = metrics.accuracy_score(y_test, y_hat)
    

print('best predict: %.2f' % min(main_accuracy))    
```
